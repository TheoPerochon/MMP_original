```python
import pandas as pd 
import os 
import numpy as np
```

```python
# read pan-cancer normalized RNA data 
df = pd.read_csv('../data/HiSeqV2_PANCAN', delimiter='\t')

# read 
hallmarks = pd.read_csv('./data_csvs/rna/metadata/hallmarks_signatures.csv').values.flatten()
hallmarks = hallmarks[~pd.isnull(hallmarks)]
hallmarks = pd.unique(hallmarks)

# filter out and transpose
df = df[df['sample'].isin(hallmarks)].reset_index(drop=True)
df = df.set_index('sample')
df = df.transpose()
df = df.sort_index(axis=1)
df = df.reset_index()
df = df.rename(columns={'index': 'sample'})
```

```python
df.to_csv('../luad_hallmarks_rna_clean.csv')
```
